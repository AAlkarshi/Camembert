```python
import pandas as pd
import matplotlib.pyplot as plt

# Lire le fichier CSV
df = pd.read_csv("camembert.csv", header=0, sep=",")

# Extraire les colonnes pertinentes pour le camembert
categories = df['Categories']
votes = df['Votes']

# Créer le camembert
plt.figure(figsize=(10, 10))
plt.pie(votes, labels=categories, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()

```
